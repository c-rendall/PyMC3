```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Introduction 

A/B testing is a part of almost every data scientist's toolbox.  Instead of calculating a point estimate or a confidence interval, A/B tests allow us to defintively answer a question of whether or not there is evidence that two values are statistically different from one another or not.  

### 2. Conversion Testing Recap

The fundamental idea of A/B testing is that we consider a hypothetical, perfectly counterfactual universe, where if the population is identical but subject to some treatment, then any differences between the populations after the study must be due to the treatment.  Let's consider two website designs, A and B.  When a user lands on our page, we randomly show them either design A or B and record the assignment.  After enough visitors have done this, we join this dataset against some metric of interest.  Let's suppose we have:

```python
# Web traffic data
visitors_to_A = 1300
visitors_to_B = 1275

conversions_from_A = 120
conversions_from_B = 125
```

We are interested in the probability of conversion, given site A or B.  We obviously want this to be as high as possible since we are a business, so the goal is to determine whether A or B has a higher probability of conversion.  To do this, we model the probability of conversion given site A or B.  Since we are modeling a probability, a good prior is the beta distribution, since it is a distribution of probabilities.  Our likelihood is a binomial distribution, since it is a distribution for the number of successes out of a given number of attempts.  Since the beta and binomial are conjugate, we don't need any MCMC!  Once we determine our values $\alpha_{0}$ and $\beta_{0}$, the posterior is simply given by Beta($\alpha_{0}$ + X, $\beta_{0}$ + N - X) where N is the number of trials and X the number of successes.  We will suppose that our prior is Beta(1, 1), which is identical to a uniform distribution on [0,1]:

```python
# Beta parameters
from scipy.stats import beta
alpha_0 = 1
beta_0 = 1

posterior_A = beta(alpha_0 + conversions_from_A, beta_0 + visitors_to_A - conversions_from_A)
posterior_B = beta(alpha_0 + conversions_from_B, beta_0 + visitors_to_B - conversions_from_B)
```

Now, we want to determine which group has a larger probability of conversion.  We can use samples from the posterior and compare the probability that samples from the posterior of A are larger than samples from the posterior of B: 

```python
# Compare the posterior of A and B
samples = 1000000
samples_posterior_A = posterior_A.rvs(samples)
samples_posterior_B = posterior_B.rvs(samples)
print(f'Probability that A converts better than B: {np.mean(samples_posterior_A > samples_posterior_B)}')
```

So the probability is about 31% for A to convert better than B, and conversely, about a 69% that B converts better than A.  This is not very significant.  Let's visualize the posterior: 

```python
# Plot the posteriors for A and B
x = np.linspace(0, 1, 500)
plt.plot(x, posterior_A.pdf(x), label='Posterior of A')
plt.plot(x, posterior_B.pdf(x), label='Posterior of B')
plt.xlim([0.05, 0.15])
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Posterior Distribution of the Conversion Rates of Web Pages $A$ and $B$')
plt.legend()
```

### 3. Adding a Linear Loss Function 

A common goal of Internet companies is to not just gain signups, but also to optimize which signup plan the user might select.  Suppose users are shown two different versions of this pricing page and we want to determine the *expected revenue* we receive per impression.  Before, we only cared about whether the user signed up or not; now, we want to know what the expected revenue to be earned is.  

In a completely transparent world, if we knew everything, we could calculate this expected value as: 

E[R] = $79_{p79} + 49_{p49} + 25_{p25} + 0_{p0}$

All of these probabilities sum to 1.  The next step is to estimate the probabilities.  We can't use a beta/binomial model for each, since the probabilities are correlated, so we need to model all the probabilities together.  This is where we use a generalization of the binomial distribution called the *multinomial* distribution.  For our signup page, the observables follow a multinomial distribution, where we don't knwo the values of the probability vector p.  

There is also a generalization of the beta distribution called the *Dirichlet* distribution.  It returns a vector of positive values that sum to 1.  The length of this vector is determined by the length of an input vector, and this input vector's values are analagous to a parameter in a prior.  

Similar to the beta/binomial conjugate pair, there is a multinomial/Dirichlet conjugate pair.  So let's suppose we have 1,000 people view our page and we have the following values:

```python
# Dirichlet posterior
from numpy.random import dirichlet
from numpy.random import multinomial

# Number of signups
N = 1000
N_79 = 10
N_49 = 46
N_25 = 80
N_0 = N - (N_79 + N_49 + N_49)

observations = np.array([N_79, N_49, N_25, N_0])
prior_params = np.array([1, 1, 1, 1])
posterior_samples = dirichlet(prior_params + observations, size=10000)

print('Two random samles from the posterior:')
print(posterior_samples[0])
print(posterior_samples[1])
```

Let's also plot the probability density of the posterior: 

```python
# Probability density of the posterior
for i, label in enumerate(['p_79', 'p_49', 'p_25', 'p_0']):
    ax = plt.hist(posterior_samples[:,i], bins=50, label=label, histtype='stepfilled')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Posterior Distributions of the Probability of Selecting Different Prices')
plt.legend()
```

We see from the plot that there is still uncertainty in what we think the probabilities might be, so there will be uncertainty in our expected value.  But that's okay! We have a posterior of our expected value, which we can get by passing each sample from the Dirichlet posterior through the `expected_revenue` function: 

```python
# Expected revenue function
def expected_revenue(P):
    return 79*P[:,0] + 49*P[:,1] + 25*P[:,2] + 0*P[:,3]

posterior_expected_revenue = expected_revenue(posterior_samples)
plt.hist(posterior_expected_revenue, histtype='stepfilled', label='Expected Revenue', bins=50)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Posterior Distributions of the Expected Revenue')
plt.legend()
```

We see from the plot that the expected revenue is somewhere around five dollars, but very unlikely to be outside four to six dollars.  

### 4. Extending to an A/B Experiment

Now let's try this same approach with two different webpages, site A and site B:

```python
# Data for pages A and B
N_A = 1000
N_A_79 = 10
N_A_49 = 46
N_A_25 = 80
N_A_0 = N_A - (N_A_79 + N_A_49 + N_A_49)
observations_A = np.array([N_A_79, N_A_49, N_A_25, N_A_0])

N_B = 2000
N_B_79 = 45
N_B_49 = 84
N_B_25 = 200
N_B_0 = N_B - (N_B_79 + N_B_49 + N_B_49)
observations_B = np.array([N_B_79, N_B_49, N_B_25, N_B_0])

prior_params = np.array([1, 1, 1, 1])

posterior_samples_A = dirichlet(prior_params + observations_A, size=10000)
posterior_samples_B = dirichlet(prior_params + observations_B, size=10000)

posterior_expected_revenue_A = expected_revenue(posterior_samples_A)
posterior_expected_revenue_B = expected_revenue(posterior_samples_B)

plt.hist(posterior_expected_revenue_A, histtype='stepfilled', label='Expected Revenue of A', bins=50)
plt.hist(posterior_expected_revenue_B, histtype='stepfilled', label='Expected Revenue of B', bins=50, alpha=0.8)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Posterior Distribution of the Expected Revenue between A and B')
plt.legend()
```

We see from the plot that the distributions for A and B are widely divergent, indicating a significant difference in how they perform.  The average expected revenue for A is about $1 less than B.  To confirm that this difference exists, we can look at the probability that the revenue for B is larger than A's: 

```python
# Calculate probability that B is larger than A
p = np.mean(posterior_expected_revenue_B > posterior_expected_revenue_A)
print(f'Probability that B has higher revenue than A: {p}')
```

This value is very high and significant, so the business should choose B going forward.  We can also plot the histogram of the differences in revenue between the posteriors of A and B: 

```python
# Difference in posteriors between A and B
posterior_diff = posterior_expected_revenue_B - posterior_expected_revenue_A
plt.hist(posterior_diff, histtype='stepfilled', label='Difference in Revenue Between A and B', bins=50)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Posterior Distribution of the Revenue Difference Between A and B')
plt.legend();
```

Looking at this posterior, there is about a 50% chance that the difference is actually *more* than one dollar, possibly even larger than two dollars.  If we are wrong about our choice of B, we probably won't lose much, since the tail of the posterior barely goes past 0.
