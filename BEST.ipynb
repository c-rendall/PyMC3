### 1. Introduction 

This notebook follows the example from `Kruschke, John. (2012) Bayesian estimation supersedes the t-test. Journal of Experimental Psychology: General`.  The default procedure for determining a difference between two groups is to use a test statistic.  Once we choose a statistic, we set up a hypothesis test to determine whether there is sufficient evidence that one group statistically is different from another.  

For all the value that classical hypothesis tests have, there are drawbacks.  There are multiple subjective choices we have to make, such as the test statistic, threshold for rejecting the null, and null hypothesis to test.  Oftentimes these choices are arbitrary, such as the 5% threshold traditionally used in many hypothesis tests.  

An arguably more informative approach for comparing two roups is one based on **estimation** rather than **testing**, driven by Bayesian probability rather than frequentist.  Rather than testing whether two groups meet some criteria to declare that they are different, we pursue an estimate of how different they are.  In this process, we also include an estimate of uncertainty in how different they are due both to our lack of complete knowledge of the model parameters and to the inherent randomness of the system.  

### 2. Drug Trial Evaluation

Suppose we are undergoing a clinical trial for drug evaluation. The trial aims to evaluate the efficacy of a “smart drug” that is supposed to increase intelligence by comparing IQ scores of individuals in a treatment arm (those receiving the drug) to those in a control arm (those recieving a placebo). There are 47 individuals and 42 individuals in the treatment and control arms, respectively.

```python
# IQ scores of individuals 
drug = (101,100,102,104,102,97,105,105,98,101,100,123,105,103,100,95,102,106,
        109,102,82,102,100,102,102,101,102,102,103,103,97,97,103,101,97,104,
        96,103,124,101,101,100,101,101,104,100,101)
placebo = (99,101,100,101,102,100,97,101,104,101,102,102,100,105,88,101,100,
           104,100,100,100,101,102,103,97,101,101,100,101,99,101,100,100,
           101,100,99,101,100,102,99,100,99)

# Convert to arrays and place in a DataFrame
y1 = np.array(drug)
y2 = np.array(placebo)
df = pd.DataFrame(
    dict(value=np.r_[y1, y2], group=np.r_[['drug'] * len(drug), ['placebo'] * len(placebo)])
)

# Histogram of values
df.hist('value', by='group');
```

Our first step in a Bayesian approach is to specify the full probability model corresponding to the problem.  A student's t-distribution is what we will choose to describe the distribution of IQ scores in each group.  For a student's t-distribution, we need to specify three parameters: a mean ($\mu$), precision, or inverse-variance ($\lambda$), and the degrees of freedom ($\nu$).  So the likelihood functions of our model are: 

$y_{i}^{(treat)}$ ~ $T(\nu, \mu_{1}, \sigma_{1})$

$y_{i}^{(placebo)}$ ~ $T(\nu, \mu_{2}, \sigma_{2})$

Since the means are real-valued, we apply normal priors on them, while arbitraily setting the hyperparameters to the pooled empirical mean of the data and twice the pooled standard deviation:

$\mu_{k}$ ~ $N(\bar{x}, 2s)$

```python
# Set up model 
mu_m = df.value.mean()
mu_s = df.value.std() * 2

with pm.Model() as model:
    group1_mean = pm.Normal('group1_mean', mu=mu_m, sd=mu_s)
    group2_mean = pm.Normal('group2_mean', mu=mu_s, sd=mu_s)
```

In order to account for the variability of IQ, we will set the group standard deviations to have a uniform distribution over a plausible range of values for the prior: 

```python
# Standard deviations with uniform prior on [1, 10]
sigma_low = 1
sigma_high = 10
with model:
    group_1_std = pm.Uniform("group_1_std", lower=sigma_low, upper=sigma_high)
    group_2_std = pm.Uniform("group_2_std", lower=sigma_low, upper=sigma_high)
```

Ok, so we have our priors over the group mean and the group standard deviations.  Now we need a prior over $\nu$, the degrees of freedom parameter.  Here, we will use an exponential distribution with a mean of 30; this ensures that there is high probability over the regions with normal to heavy-tailed data under a student-t distribution. 

```python
# Prior for nu
with model:
    nu = pm.Exponential('nu_minus_one', 1 / 29.0) + 1
    
pm.kdeplot(np.random.exponential(30, size=10000), fill_kwargs={"alpha": 0.5});
```

PyMC3 parametrizes the student-t distribution in terms of precision rather than standard deviation, so we need to transform the standard deviation before specifying the likelihoods:

```python
# Transform standard deviations to precision 
with model: 
    lambda1 = group_1_std ** -2
    lambda2 = group_2_std ** -2
    
    group1 = pm.StudentT('drug', nu=nu, mu=group1_mean, lam=lambda1, observed=y1)
    group2 = pm.StudentT('placebo', nu=nu, mu=group2_mean, lam=lambda2, observed=y2)
```

Great! So we have now fully specified our probabilistic model.  We can now specify deterministic nodes in our model for the difference between the group means and the difference between the group standard deviations.  We wrap them in `Deterministic()` to tell PyMC3 we want to retain the values.  We will also record the effect size, which is the difference in means scaled by the pooled estimates of standard deviation.  

```python
# Deterministic values
with model:
    diff_of_means = pm.Deterministic('difference of means', group1_mean - group2_mean)
    diff_of_stds = pm.Deterministic('difference of SDs', group_1_std - group_2_std)
    effect_size = pm.Deterministic('effect size', 
                                   diff_of_means / np.sqrt((group_1_std ** 2 + group_2_std **2) / 2)
                                  )
```

Now we can fit the model and evaluate the output: 

```python
# Fit model
with model:
    trace = pm.sample(2000)
```

We can now plot the stochastic parameters of the model.  We can use `plot_posterior` to show the 95% credible interval and posterior mean.  

```python
# Plot posterior 
pm.plot_posterior(
    trace,
    var_names=["group1_mean", "group2_mean", "group_1_std", "group_2_std", "nu_minus_one"],
    color="#87ceeb",
);
```

Looking at these differences, we can conclude that there are meaningful differences between the three parameters for both groups.  These estimates suggest that the "smart drug" increased both the expected scores, and also the variability in scores across the sample.  The effect size and difference in standard deviations is similarly positive.  Therefore, this does not rule out that some receipients may be adversely affected by the drug, while others benefit from it.  

```python
# Plot posterior for the difference of means, standard deviations, and effect size
pm.plot_posterior(
    trace,
    var_names=["difference of means", "difference of SDs", "effect size"],
    ref_val=0,
    color="#87ceeb",
);
```

```python
# Show finaly summary
pm.summary(trace, varnames=["difference of means", "difference of SDs", "effect size"])
```
